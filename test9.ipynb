```python
import re

# Example list of time strings
time = ['4.2', '18.5', '2:26.6', '13:22.7', '137:11.5', '1985:48.2']

# Function to convert time string to seconds
def time_to_seconds(time_str):
    if re.match(r'^\d+(\.\d+)?$', time_str):  # Matches 'SS.s' format
        return float(time_str)
    elif re.match(r'^\d+:\d+(\.\d+)?$', time_str):  # Matches 'MM:SS.s' format
        minutes, seconds = map(float, time_str.split(':'))
        return minutes * 60 + seconds
    elif re.match(r'^\d+:\d+:\d+(\.\d+)?$', time_str):  # Matches 'HH:MM:SS.s' format
        hours, minutes, seconds = map(float, time_str.split(':'))
        return hours * 3600 + minutes * 60 + seconds
    else:
        raise ValueError(f"Invalid time format: {time_str}")

# Convert each element in the time list to seconds
time_in_seconds = [time_to_seconds(time_str) for time_str in time]

print(time_in_seconds)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the SCAD penalty function
def scad_penalty(beta, lambda_, a=3.7):
    penalty = np.zeros_like(beta)
    abs_beta = np.abs(beta)
    
    # Case 1: |beta| <= lambda
    mask1 = abs_beta <= lambda_
    penalty[mask1] = lambda_ * abs_beta[mask1]
    
    # Case 2: lambda < |beta| <= a * lambda
    mask2 = (abs_beta > lambda_) & (abs_beta <= a * lambda_)
    penalty[mask2] = (-abs_beta[mask2]**2 + 2 * a * lambda_ * abs_beta[mask2] - lambda_**2) / (2 * (a - 1))
    
    # Case 3: |beta| > a * lambda
    mask3 = abs_beta > a * lambda_
    penalty[mask3] = (a + 1) * lambda_**2 / 2
    
    return penalty

# Define the range of beta values
beta = np.linspace(-10, 10, 400)

# Define the lambda parameter
lambda_ = 2

# Calculate the SCAD penalty for each beta value
penalty = scad_penalty(beta, lambda_)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(beta, penalty, label='SCAD Penalty')
plt.xlabel('x', fontsize=14, fontweight='bold')
plt.ylabel('Penalty', fontsize=14, fontweight='bold')
plt.title('SCAD Penalty Function', fontsize=16, fontweight='bold')
plt.legend()
plt.grid(True)
plt.show()
```
